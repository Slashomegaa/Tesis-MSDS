```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
Data = pd.read_excel('Government Quality Index Proxy.xlsx')
```

```python
Data.columns
```

Se seleccionan todas las variables

```python
df = Data[['Government Effectiveness: Estimate','Voice and Accountability: Estimate', 'Regulatory Quality: Estimate',
       'Rule of Law: Estimate', 'Control of Corruption: Estimate',
       'Political Stability and Absence of Violence/Terrorism: Estimate']]
```

```python
df
```

```python
scaler = StandardScaler()
df_gqi_scaled = scaler.fit_transform(df)
```

```python
pca = PCA(n_components=None)
pca.fit(df_gqi_scaled)
```

```python
explained_variances = pca.explained_variance_ratio_
print("Varianza explicada por cada componente:", explained_variances)
print("Varianza explicada acumulada:", np.cumsum(explained_variances))
```

```python
pca_scores = pca.transform(df_gqi_scaled)
```

```python
ict_index_pc1 = pca_scores[:, 0]
```

```python
df['GQI_mean'] = df_gqi_scaled.mean(axis=1)
```

```python
df.loc[df.dropna().index, 'GQI_Index'] = ict_index_pc1
```

```python
df
```

```python
df.to_csv("GQI.csv", index=False)
```
