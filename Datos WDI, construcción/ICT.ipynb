```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
Data = pd.read_excel('ICT Proxy Data.xlsx')
```

```python
Data.columns
```

Seleccionaremos 'Fixed broadband subscriptions (per 100 people)', 'Mobile cellular subscriptions (per 100 people)', 'ICT goods exports (% of GDP)', 'ICT goods imports (% of GDP)' y 'Individuals using the Internet (% of population)'

```python
df = Data[['Fixed broadband subscriptions (per 100 people)','Mobile cellular subscriptions (per 100 people)','ICT goods exports (% of GDP)','ICT goods imports (% of GDP)','Individuals using the Internet (% of population)']]
```

```python
df
```

```python
scaler = StandardScaler()
df_ict_scaled = scaler.fit_transform(df)
```

```python
pca = PCA(n_components=None)
pca.fit(df_ict_scaled)
```

```python
explained_variances = pca.explained_variance_ratio_
print("Varianza explicada por cada componente:", explained_variances)
print("Varianza explicada acumulada:", np.cumsum(explained_variances))
```

```python
pca_scores = pca.transform(df_ict_scaled)
```

```python
ict_index_pc1 = pca_scores[:, 0]
```

Agregamos mean

```python
df['ICT_mean'] = df_ict_scaled.mean(axis=1)
```

Agregamos ICT PC1

```python
df.loc[df.dropna().index, 'ICT_Index'] = ict_index_pc1
```
