```python
def check_stats_diff(HS_92, Prod_trade):
    years = HS_92['year'].unique()
    usecols = ['country_iso3_code', 'product_hs92_code', 'year', 'export_value']
    HS_92 = HS_92[usecols]
    Prod_trade = Prod_trade[usecols]

    HS_92_grouped = HS_92.groupby(['country_iso3_code', 'product_hs92_code', 'year'], observed=True)['export_value'].sum().reset_index()
    #Guardaremos las estadisticas en un dataframe
    stats_df_total = pd.DataFrame(columns=['year', 'mean_diff', 'median_diff', 'max_diff', 'min_diff'])

    for year in years:
        Prod_trade_year = Prod_trade[Prod_trade['year'] == year]
        HS_92_year = HS_92_grouped[HS_92_grouped['year'] == year]
        diff_df = HS_92_year.merge(Prod_trade_year, on=['country_iso3_code', 'product_hs92_code', 'year'], suffixes=('_hs92', '_prod'), how='outer', indicator=True)
        #drop NaN values in export_value columns
        diff_df['export_value_hs92'] = diff_df['export_value_hs92'].dropna()
        diff_df['export_value_prod'] = diff_df['export_value_prod'].dropna()
        #Select only with merge indicator 'both'
        diff_df = diff_df[diff_df['_merge'] == 'both']
        #Calculate the difference between export_value_prod and export_value_h
        # Absolute difference
        diff_df['export_value_diff'] = (diff_df['export_value_prod'] - diff_df['export_value_hs92']).abs()
        diff_df['export_value_diff'] = diff_df['export_value_diff'].astype('int64')
        stats = {
        'mean_diff': diff_df['export_value_diff'].mean(),
        'median_diff': diff_df['export_value_diff'].median(),
        'max_diff': diff_df['export_value_diff'].max(),
        'min_diff': diff_df['export_value_diff'].min()
        }
        stats['year'] = year
        stats_df_total = pd.concat([stats_df_total, pd.DataFrame([stats])], ignore_index=True)
    #Regresamos stats_df_total ordenado por year
    stats_df_total = stats_df_total.sort_values(by='year').reset_index(drop=True)
    return stats_df_total

    
```

```python
check_stats_diff(HS_92_0009, Prod_trade)
```

```python
HS_92_0009_grouped = HS_92_0009.groupby(['country_iso3_code', 'partner_iso3_code', 'product_hs92_code', 'year'], observed=True)[['export_value', 'import_value']].sum().reset_index()
```
